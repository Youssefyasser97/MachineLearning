```python
# -*- coding: utf-8 -*-
"""
Created on Thu Oct 24 17:54:50 2019

@author: w
"""
from PIL import Image
import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt



Xdelta=np.zeros([2400,785])
x = [ i for i in range(1,2401)]
x1=np.ones([2400])
filelist =list(map(str,x))
t0=np.zeros([2400])

x2 = [ i for i in range(1,201)]

filelist1 =list(map(str,x2))

XdeltaTest=np.zeros([200,785])



            




#putting training data into matrix

for imagefile in filelist:
    im=Image.open("F:\Semester 9\ML\Assignment2\Train/"+imagefile+".JPG")
    arrayimage =im.convert("L")
    imgarray = np.array(arrayimage)
    imgarray=imgarray.reshape(28*28)
    Xdelta[int(imagefile)-1][0:784]=imgarray
    Xdelta[int(imagefile)-1][784]=1

            
                    
t1=np.zeros([2400])
t2=np.zeros([2400])
t3=np.zeros([2400])
t4=np.zeros([2400])
t5=np.zeros([2400])
t6=np.zeros([2400])
t7 =np.zeros([2400])
t8=np.zeros([2400])
t9=np.zeros([2400])                    

t0[0:240]=1
t0[240:2400]=-1

t1[0:240]=-1
t1[240:480]=1
t1[480:2400]=-1

t2[0:480]=-1
t2[480:720]=1
t2[720:2400]=-1

t3[0:720]=-1
t3[720:960]=1
t3[960:2400]=-1

t4[0:960]=-1
t4[960:1200]=1
t4[1200:2400]=-1



t5[0:1200]=-1
t5[1200:1440]=1
t5[1440:2400]=-1

t6[0:1440]=-1
t6[1440:1680]=1
t6[1680:2400]=-1

t7[0:1680]=-1
t7[1680:1920]=1
t7[1920:2400]=-1

t8[0:1920]=-1
t8[1920:2160]=1
t8[2160:2400]=-1

t9[0:2160]=-1
t9[2160:2400]=1

T=np.vstack((t0,t1,t2,t3,t4,t5,t6,t7,t8,t9))

for imagefile1 in filelist1:
    im1=Image.open("F:\Semester 9\ML\Assignment2\Test/"+imagefile1+".JPG")
    arrayimage1 =im1.convert("L")
    imgarray1 = np.array(arrayimage1)
    imgarray1=imgarray1.reshape(28*28)
    XdeltaTest[int(imagefile1)-1][0:784]=imgarray1
    XdeltaTest[int(imagefile1)-1][784]=1

with open("F:\Semester 9\ML\Assignment2\Test/TestLabels.txt", "r") as f:
    TestLabels = [int(x) for x in f.read().split()]



    

for n in [1,10**-1,10**-2,10**-3,10**-4,10**-5,10**-6,10**-7,10**-8,10**-9]:
    ConfusionMatrix=np.zeros([10,10])
    finalMatrix=np.zeros([200,10])
    W=np.zeros([10,785])
    predictions=np.zeros([200])
    for i in range(0,10):
        W[i][0]=1
    
    for c in range(0,10):
        for epoch in range(0,500):
            for img in range(0,2400):
                if ((W[c].dot(Xdelta[img]))>=0):
                    result=1
                else:
                    result=-1
                if(result!=T[c][img]):
                    W[c]+= (n*Xdelta[img]*T[c][img])
                    
                
    # #Y(x)=WT*X
    for h in range (0,200):
        finalMatrix[h]=W.dot(XdeltaTest[h])
    
    predictions=np.argmax(finalMatrix,axis=1)
    

    # #Creating Confusion Matrix
    for k in range(0,200):
        ConfusionMatrix[TestLabels[k]][predictions[k]]+=1

    fig=plt.figure(6)
    sns.heatmap(ConfusionMatrix,annot= True)

    plt.title('Confusion Matrix')

    plt.xlabel('Predicted Class')
    plt.ylabel('Actual Class')
    confusion="confusion"+"-"+str(abs(n))+".jpg"
    plt.show()
    fig.savefig("F:\Semester 9\ML\Assignment2/"+confusion)
                  




```
