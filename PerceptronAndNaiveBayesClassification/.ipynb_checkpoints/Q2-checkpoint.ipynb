```python
import math
from PIL import Image
import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt
Xdelta=np.zeros([2400,784])



x = [ i for i in range(1,2401)]
x1=np.ones([2400])
filelist =list(map(str,x))
t0=np.zeros([2400])

x2 = [ i for i in range(1,201)]

filelist1 =list(map(str,x2))

XdeltaTest=np.zeros([200,784])


#putting training data into matrix

for imagefile in filelist:
    im=Image.open("E:\Semester 9\ML\Assignment2\Train/"+imagefile+".JPG")
    arrayimage =im.convert("L")
    imgarray = np.array(arrayimage)
    imgarray=imgarray.reshape(28*28)
    Xdelta[int(imagefile)-1][0:784]=imgarray
    

            
                    
t1=np.zeros([2400])
t2=np.zeros([2400])
t3=np.zeros([2400])
t4=np.zeros([2400])
t5=np.zeros([2400])
t6=np.zeros([2400])
t7 =np.zeros([2400])
t8=np.zeros([2400])
t9=np.zeros([2400])                    

t0[0:240]=1
t0[240:2400]=-1

t1[0:240]=-1
t1[240:480]=1
t1[480:2400]=-1

t2[0:480]=-1
t2[480:720]=1
t2[720:2400]=-1

t3[0:720]=-1
t3[720:960]=1
t3[960:2400]=-1

t4[0:960]=-1
t4[960:1200]=1
t4[1200:2400]=-1



t5[0:1200]=-1
t5[1200:1440]=1
t5[1440:2400]=-1

t6[0:1440]=-1
t6[1440:1680]=1
t6[1680:2400]=-1

t7[0:1680]=-1
t7[1680:1920]=1
t7[1920:2400]=-1

t8[0:1920]=-1
t8[1920:2160]=1
t8[2160:2400]=-1

t9[0:2160]=-1
t9[2160:2400]=1

T=np.vstack((t0,t1,t2,t3,t4,t5,t6,t7,t8,t9))


        


for imagefile1 in filelist1:
    im1=Image.open("E:\Semester 9\ML\Assignment2\Test/"+imagefile1+".JPG")
    arrayimage1 =im1.convert("L")
    imgarray1 = np.array(arrayimage1)
    imgarray1=imgarray1.reshape(28*28)
    XdeltaTest[int(imagefile1)-1][0:784]=imgarray1
    

with open("E:\Semester 9\ML\Assignment2\Test/TestLabels.txt", "r") as f:
    TestLabels = [int(x) for x in f.read().split()]

for g1 in range(0,2400):
    for k1 in range(0,784):
        Xdelta[g1][k1]=Xdelta[g1][k1]/255

reshapedmatrix=Xdelta.reshape(10,240,784)
mean=np.zeros([10,784])
variance=np.zeros([10,784])

for i in range(0,10):
    mean[i]=np.mean(reshapedmatrix[i],axis=0)
    variance[i]=np.var(reshapedmatrix[i],axis=0)



for g in range(0,10):
    for k in range(0,784):
        if (variance[g][k]<0.01):
            variance[g][k]=0.01
            
for k in range(0,200):
    for f in range(0,784):
        XdeltaTest[k][f]=XdeltaTest[k][f]/255

finalMatrix=np.ones([200,10,784])
 
for i in range(0,len(XdeltaTest)):
    for j in range(0,784):
        for k in range(0,10):
            
            exp1=(1/   math.sqrt(2*(math.pi*(variance[k][j]))))
            exp2=(-1*  ( (XdeltaTest[i][j])-(mean[k][j]) )**2  )
            exp3= (2*(variance[k][j])) 
            exp4=exp2/exp3
            exp5=math.exp(exp4)
            finalMatrix[i][k][j]=exp1*exp5
           
        
finalMatrix1=np.ones([200,10])
for i in range(0,200):
    for j in range(0,10):
        finalMatrix1[i]=np.prod(finalMatrix[i],axis=1)




ConfusionMatrix=np.zeros([10,10])
predictions=np.argmax(finalMatrix1,axis=1)


#Creating Confusion Matrix
for k in range(0,200):
    
    ConfusionMatrix[TestLabels[k]][predictions[k]]+=1



fig=plt.figure(6)
sns.heatmap(ConfusionMatrix,annot= True)


plt.title('Confusion Matrix')

plt.xlabel('Predicted Class')
plt.ylabel('Actual Class')


plt.show()
fig.savefig("E:\Semester 9\ML\Assignment2/confusionQ2.jpg")
```
